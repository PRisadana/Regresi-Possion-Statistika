# Cleansing Data Youtube Rewind 2023

#### Import Library

```python
from dotenv import load_dotenv
from pymongo import MongoClient
import os
import pandas as pd
```

#### Import url dari .env

```python
# Memuat value dari file .env
load_dotenv()

mongodb_url = os.getenv('URL_SANDY')
local_url = os.getenv('URL_LOCAL')
```

```python
database_name = "regresiPossionAnalisisSentimenYTRewind"
collection_from = "collectData2023"
collection_name = "cleanedData2023"
```

#### Import dataset dari MongoDB

```python
# Koneksi ke MongoDB
client = MongoClient(local_url)

# Pilih database dan koleksi
db = client[database_name]
collection = db[collection_from]

# Query data dari koleksi
cursor = collection.find()

# Konversi cursor ke list dan kemudian ke DataFrame
data = list(cursor)
df = pd.DataFrame(data)

# Menutup koneksi
client.close()
```

```python
df
```

#### Menghapus data text yang duplikat

```python
# Mengonversi semua nilai dalam kolom 'textOriginal' menjadi string
df['textOriginal'] = df['textOriginal'].apply(str)

# Menghapus duplikat berdasarkan kolom 'textOriginal'
df = df.drop_duplicates(subset=['textOriginal'])

# Melihat jumlah data yang duplikat
num_duplicates = df.duplicated(subset=['textOriginal']).sum()
print(f"Jumlah duplikat setelah penghapusan: {num_duplicates}")

# Melihat bentuk DataFrame setelah penghapusan duplikat
print(df.shape)
```

```python
df
```

#### Menghapus data author yang duplikat

```python
df = df.drop_duplicates(subset=['authorChannelUrl'])

df.shape
```

#### Menghilangkan atribut yang tidak diperlukan

```python
# Memeriksa kolom yang tersedia
print(df.columns)

# Membuat daftar kolom yang ingin disertakan, kecuali 'parentID'
columns_to_include = [col for col in df.columns if col != 'parentId']

# Membuat DataFrame baru dengan kolom yang disertakan
df = df[columns_to_include]

# Menampilkan DataFrame
print(df.head())
```

#### Menghapus nilai yang kosong

```python
df = df.dropna()

# Melihat data yang kosong
df.isnull().sum()
```

```python
df
```

Dari yang awalnya berjumlah 133687 row berkurang menjadi 99235 row

#### Menyimpan Clean Comments ke MongoDB Local

```python
# Koneksi ke MongoDB
client = MongoClient(local_url)
db = client[database_name]
collection = db[collection_name]

# Konversi DataFrame ke bentuk dictionary
records = df.to_dict(orient='records')

# Menyisipkan data ke MongoDB
collection.insert_many(records)

print("Data berhasil diimpor ke MongoDB Local.")
```
